```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### Take 2 images 1st - dog , 2nd - horse  and generate two new images having dog contains horse head and vice versa

```python
cat =cv2.imread('catt.jpg')
horse = cv2.imread('horsee.jpg')
cat = cv2.cvtColor(cat, cv2.COLOR_BGR2RGB)
horse = cv2.cvtColor(horse, cv2.COLOR_BGR2RGB)
plt.imshow(cat)
plt.show()
plt.imshow(horse)
```

```python
cat_face = cat[300:800,450:1000]
horse_face= horse[0:600,0:250]
horse_face_resize=cv2.resize(horse_face,(550,500))
cat_face_resize=cv2.resize(cat_face,(250,600))

plt.imshow(cat_face_resize)
plt.show()
plt.imshow(horse_face_resize)
```

```python
cat[300:800,450:1000]=horse_face_resize
horse[0:600,0:250] =cat_face_resize
plt.imshow(horse)
plt.show()
plt.imshow(cat)
```
